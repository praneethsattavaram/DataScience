```python
data['Winning bid'] = data['Winning bid'].astype('float')
```

```python
null_count = data.isnull().sum()
print(null_count)
```

```python

data['Base price'] = data['Base price'].astype('object')
```

```python
data['premium'] = data['Winning bid'] - data['Base price']
```

```python
data.head()
```

```python
data['Base price'] = data['Base price'].astype('float64')
```

```python
data.groupby('Year')['Base price'].mean()
```

```python
baseprice_by_year = data.groupby('Year')['Base price'].mean().reset_index()
print(baseprice_by_year)
```

```python
baseprice_by_team_year = data.groupby(['Year', 'Team']) ['Base price'].mean().reset_index
```

```python
print(baseprice_by_team_year)
```

```python
baseprice_comparison = baseprice_by_team_year.merge(baseprice_by_year, on = 'Year', how = 'outer')
```
